```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Specify the directory containing your CSV files
csv_directory = r'Data'  # Replace with the actual path to your CSV files

# Get a list of all CSV files in the specified directory
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

# Create an empty dictionary to store DataFrames
df = {}

# Loop through each CSV file and reading it into a DataFrame
for csv_file in csv_files:
    # Create a key for the DataFrame using the file name (without the extension)
    key = os.path.splitext(csv_file)[0]
    
    # Construct the full path to the CSV file
    csv_file_path = os.path.join(csv_directory, csv_file)
    
    # Read the CSV file into a DataFrame
    df[key] = pd.read_csv(csv_file_path)

df.keys()

# Display the imported DataFrames


  #reading the csv file into the pandas dataframe "df"
#df_2010=pd.read_csv(r'C:\Users\bhoga\Documents\Project\Data')
  #displaying the first 5 rows of the pandas dataframe using the head function
#print(df_2010.head(5))

  #creating a copy of the pandas dataframe as to retain the original dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Ensure DATE is in datetime format
df_daily['DATE'] = pd.to_datetime(df_daily['DATE'])

# Create a new column 'Day_Number' to represent each date as a numerical value
df_daily['Day_Number'] = (df_daily['DATE'] - df_daily['DATE'].min()).dt.days

# Define the feature (X) and the target (y)
X = df_daily[['Day_Number']]  # Features
y = df_daily['TEMP AVG']      # Target variable
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create linear regression object
model = LinearRegression()

# Train the model using the training sets
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Predicting the Test set results
y_pred = model.predict(X_test)
# Calculating the metrics
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the results
print(f"R-squared: {r_squared:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
future_years = 1
last_date = df_daily['DATE'].max()
future_dates = pd.date_range(start=last_date, periods=future_years * 365, freq='D')[1:]

# Convert future dates to 'Day_Number'
future_day_numbers = (future_dates - df_daily['DATE'].min()).days

# Predicting temperatures
future_temps = model.predict(np.array(future_day_numbers).reshape(-1, 1))

# Creating a DataFrame for the forecast
forecast_df = pd.DataFrame({'DATE': future_dates, 'Predicted_TEMP_AVG': future_temps})
```

```python
# Scatter plot of the forecasted data
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DATE', y='Predicted_TEMP_AVG', data=forecast_df)
plt.xlabel('Date')
plt.ylabel('Predicted Average Temperature')
plt.title('Scatter Plot of Forecasted Average Temperatures')
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

# Ensure DATE is in datetime format and set as index
df_daily['DATE'] = pd.to_datetime(df_daily['DATE'])
df_daily.set_index('DATE', inplace=True)

# Resample data to daily frequency if it's not already
df_daily = df_daily.resample('D').mean()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
result = adfuller(df_daily['TEMP AVG'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

# Plot autocorrelation
autocorrelation_plot(df_daily['TEMP AVG'])
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Example: ARIMA with some parameters (p, d, q)
model = ARIMA(df_daily['TEMP AVG'], order=(5,1,0))  # Change the order based on your data
model_fit = model.fit()
```

```python
# Forecasting the next 365 days (1 year)
forecast = model_fit.forecast(steps=15)
forecast_dates = pd.date_range(start=df_daily.index[-5], periods=25, freq='D')[1:]

forecast_series = pd.Series(forecast, index=forecast_dates)
forecast_series.plot()
plt.title('next month Temperature Forecast')
plt.show()
```
